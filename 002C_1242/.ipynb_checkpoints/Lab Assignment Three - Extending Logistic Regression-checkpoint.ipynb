**<u>Pre-processing Methods (dimensionality reduction, scaling, etc)</u>**

Determine the number of rows containing NaN values in the dataset and consider the potential removal of these rows.


```python
def count_missing_values(data):
    missing_values_count = data.isnull().sum().sum()
    return missing_values_count
```

```python
def handle_missing_values(X, Y):
    """
    Removing rows with missing values.
    """
    missing_values_count_X = X.isnull().sum()
    missing_values_count_Y = Y.isnull().sum()

    if missing_values_count_X.sum() > 0 or missing_values_count_Y > 0:
        print("Warning: Input data contains missing values!")
        if missing_values_count_X.sum() > 0:
            print(f"Missing values in X: {missing_values_count_X.sum()}")
        if missing_values_count_Y > 0:
            print(f"Missing values in Y: {missing_values_count_Y}")
        print()

        # Remove rows with missing values
        cleaned_data = X.dropna()
        Y_cleaned = Y.loc[cleaned_data.index]

        # Reset index
        X_cleaned = cleaned_data.reset_index(drop=True)
        Y_cleaned = Y_cleaned.reset_index(drop=True)
    else:
        X_cleaned = X
        Y_cleaned = Y

    return X_cleaned, Y_cleaned

X_cleaned, Y_cleaned = handle_missing_values(X, Y)
```

```python
def evaluate_model(X, Y, label='Dataset', test_size=0.33, random_state=33, cv=5):
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=random_state)
    model = RandomForestClassifier(random_state=random_state)
    scores = cross_val_score(model, X_train, y_train, cv=cv)
    
    print(f"{label} Cross-Validation Scores:", scores)
    print(f"{label} Mean Score:", scores.mean())
    print()

    return X_train, X_test, y_train, y_test
```

# Modeling (5 points total)

**<u>The implementation of logistic regression must be written only from the examples given to you by the instructor. </u>** No credit will be assigned to teams that copy implementations from another source, regardless of if the code is properly cited. 

**<u>[2 points] Create a custom, one-versus-all logistic regression classifier using numpy and scipy to optimize.</u>** Use object oriented conventions identical to scikit-learn. You should start with the template developed by the instructor in the course. You should add the following functionality to the logistic regression classifier:
- Ability to choose optimization technique when class is instantiated: either steepest ascent, stochastic gradient ascent, and Newton's method. It is recommended to call this the "solver" input for the class.
- Update the gradient calculation to include a customizable regularization term (either using no regularization, L1 regularization, L2 regularization, or both L1 and L2 regularization). Associate a cost with the regularization term, "C", that can be adjusted when the class is instantiated.  

```python
class BinaryLogisticRegressionBase:
    # private:
    def __init__(self, eta, iterations=20):
        self.eta = eta
        self.iters = iterations
        # internally we will store the weights as self.w_ to keep with sklearn conventions
    
    def __str__(self):
        return 'Base Binary Logistic Regression Object, Not Trainable'
    
    # convenience, private and static:
    @staticmethod
    def _sigmoid(theta):
        return 1/(1+np.exp(-theta)) 
    
    @staticmethod
    def _add_intercept(X):
        return np.hstack((np.ones((X.shape[0],1)),X)) # add bias term
    
    # public:
    def predict_proba(self, X, add_intercept=True):
        # add bias term if requested
        Xb = self._add_intercept(X) if add_intercept else X
        
        # return self._sigmoid(Xb @ self.w_) # return the probability y=1
        probabilities = self._sigmoid(Xb @ self.w_)
        if probabilities.ndim == 1:  # make probabilities a two-dimensional array
            probabilities = probabilities.reshape(-1, 1)
        return probabilities # return the probability y=1
    
    def predict(self,X):
        return (self.predict_proba(X)>0.5) #return the actual prediction
```

```python
class BinaryLogisticRegression(BinaryLogisticRegressionBase):
    def __init__(self, eta, iterations=20, solver=None, regularization=None, c=None):
        super().__init__(eta, iterations)
        self.solver = solver
        self.regularization = regularization
        self.c = c
    
    #private:
    def __str__(self):
        if(hasattr(self,'w_')):
            return 'Binary Logistic Regression Object with coefficients:\n'+ str(self.w_) # is we have trained the object
        else:
            return 'Untrained Binary Logistic Regression Object'

    def _add_regularization(self, gradient):
        if self.penalty == 'none':
            return gradient

        # regularization_term >> bias_correction >> regularization_with_bias >> updated_gradient
        elif self.penalty == 'l1':
            return gradient + self.c * np.vstack([np.zeros((1, 1)), np.sign(self.w_[1:])])
            
        # bias_correction >> weights_without_bias >> weights_with_bias >> regularization_term >> updated_gradient
        elif self.penalty == 'l2':
            return gradient + self.c * np.vstack([np.zeros((1, 1)), self.w_[1:]])
            
        # l1_component >> l2_component >> regularization_term >> bias_correction >> regularization_with_bias >> updated_gradient
        elif self.penalty == 'l1l2':
            l1_component = np.sign(self.w_[1:])
            l2_component = self.w_[1:]
            return gradient + self.C * np.vstack([np.zeros((1, 1)), 0.5 * l1_component + 0.5 * l2_component])
            
        else:
            raise ValueError("Invalid penalty.")
        
    def _get_gradient(self, X, y):
        ydiff = y - self.predict_proba(X, add_intercept=False).ravel()
        gradient = np.mean(X * ydiff[:, np.newaxis], axis=0)
        gradient = gradient.reshape(self.w_.shape)
        return self._add_regularization(gradient)

    def _steepest_ascent(self, X, y):
        for _ in range(self.iters):
            gradient = self._get_gradient(X, y)
            self.w_ += self.eta * gradient

    def _stochastic_gradient_ascent(self, X, y):
        for _ in range(self.iters):
            for i in range(X.shape[0]):
                # single sample for gradient calculation
                # X_single_sample >> y_single_sample >> gradient
                gradient = self._get_gradient(X[i:i+1], y[i:i+1])
                self.w_ += self.eta * gradient

    def _newtons_method(self, X, y):
        def objective_func(w, X, y, c):
            m = X.shape[0]
            y_pred = self._sigmoid(np.dot(X, w))
            # positive_class_term >> negative_class_term
            cross_entropy_loss = -np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
            if self.regularization == 'l2':
                # l2_regularization_term
                cross_entropy_loss += (self.c / 2) * np.sum(np.square(w[1:]))
            return cross_entropy_loss

        def gradient_func(w, X, y, c):
            y_pred = self._sigmoid(np.dot(X, w))
            gradient = np.dot(X.T, y_pred - y) / y.size
            if self.regularization == 'l2':
                gradient[1:] += c * w[1:] / y.size
            return gradient

        self.w_, _, _ = fmin_tnc(func=objective_func, x0=self.w_, fprime=gradient_func, args=(X, y, self.c))

       
    # public:
    def fit(self, X, y):
        Xb = self._add_intercept(X)
        num_samples, num_features = Xb.shape
        self.w_ = np.zeros((num_features, 1))
        
        if self.solver == 'steepest':
            self._steepest_ascent(Xb, y)
        elif self.solver == 'stochastic':
            self._stochastic_gradient_ascent(Xb, y)
        elif self.solver == 'newton':
            self._newtons_method(Xb, y)
        else:
            raise ValueError("Invalid solver.")
```

```python
class VectorBinaryLogisticRegression(BinaryLogisticRegression):
    # inherit from our previous class to get same functionality
    @staticmethod
    def _sigmoid(theta):
        # increase stability, redefine sigmoid operation
        return expit(theta) #1/(1+np.exp(-theta))
    
    # but overwrite the gradient calculation
    def _get_gradient(self,X,y):
        ydiff = y-self.predict_proba(X,add_intercept=False).ravel() # get y difference
        # gradient = np.mean(X * ydiff[:,np.newaxis], axis=0) # make ydiff a column vector and multiply through
        gradient = np.mean(X * ydiff.values[:,np.newaxis], axis=0) # make ydiff a column vector and multiply through

        return gradient.reshape(self.w_.shape)
```

```python
class LogisticRegression:
    def __init__(self, eta, iterations=20, solver=None, regularization=None, c=None):
        self.eta = eta
        self.iters = iterations
        # internally we will store the weights as self.w_ to keep with sklearn conventions
        self.solver = solver
        self.regularization = regularization
        self.c = c
    
    def __str__(self):
        if(hasattr(self,'w_')):
            return 'MultiClass Logistic Regression Object with coefficients:\n'+ str(self.w_) # is we have trained the object
        else:
            return 'Untrained MultiClass Logistic Regression Object'
        
    def fit(self,X,y):
        num_samples, num_features = X.shape
        self.unique_ = np.unique(y) # get each unique class value
        num_unique_classes = len(self.unique_)
        self.classifiers_ = [] # will fill this array with binary classifiers
        
        for i,yval in enumerate(self.unique_): # for each unique value
            y_binary = (y==yval) # create a binary problem
            # train the binary classifier for this class
            blr = VectorBinaryLogisticRegression(self.eta,
                                                 self.iters,
                                                 self.solver,
                                                 self.regularization,
                                                 self.c)
            blr.fit(X,y_binary)
            # add the trained classifier to the list
            self.classifiers_.append(blr)
            
        # save all the weights into one matrix, separate column for each class
        self.w_ = np.hstack([x.w_ for x in self.classifiers_]).T
        
    def predict_proba(self,X):
        probs = []
        for blr in self.classifiers_:
            probs.append(blr.predict_proba(X)) # get probability for each classifier
        
        return np.hstack(probs) # make into single matrix
    
    def predict(self,X):
        return self.unique_[np.argmax(self.predict_proba(X),axis=1)] # take argmax along row
```

```python
def lr_optimizer(solvers, regularizations, c_values):
    results = []
    for solver in solvers:
        for regularization in regularizations:
            for c in c_values:
                lr = LogisticRegression(0.1, 500, solver, regularization, c)
                lr.fit(X_train, y_train)
                yhat = lr.predict(X_test)
                accuracy = accuracy_score(y_test, yhat)
                results.append((solver, regularization, c, accuracy))
                print(results)
                print(f'Count: {len(results)}')
                print('-----')
    return results
```

```python
def check_valid_combination(solver, penalty):
    if solver == 'newton-cg' and penalty == 'l1':
        return False
    if solver == 'sag' and penalty != 'l2' and penalty != 'none':
        return False
    if solver == 'newton-cg' and penalty == 'elasticnet':
        return False
    return True
```

```python
param_grid = {
    'solver': ['saga', 'newton-cg', 'sag'],  
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],   
    'C': [0.0001, 1000, 10000],
    'l1_ratio': [0.1, 0.5, 0.9]  # Elastic Net specific parameter
}

def evaluate_logistic_regression(param_grid, X_train, X_test, y_train, y_test):
    results = []

    # Iterate over parameter combinations
    for solver in param_grid['solver']:
        for penalty in param_grid['penalty']:
            for C in param_grid['C']:
                if not check_valid_combination(solver, penalty):
                    continue

                if penalty == 'elasticnet':
                    for l1_ratio in param_grid['l1_ratio']:
                        lr = SKLogisticRegression(solver=solver, penalty=penalty, C=C, l1_ratio=l1_ratio, max_iter=500)
                        lr.fit(X_train, y_train)
                        y_pred = lr.predict(X_test)
                        accuracy = accuracy_score(y_test, y_pred)
                        results.append({
                            'solver': solver,
                            'penalty': penalty,
                            'C': C,
                            'l1_ratio': l1_ratio,
                            'accuracy': accuracy
                        })
                else:
                    lr = SKLogisticRegression(solver=solver, penalty=penalty, C=C, max_iter=500)
                    lr.fit(X_train, y_train)
                    y_pred = lr.predict(X_test)
                    accuracy = accuracy_score(y_test, y_pred)
                    results.append({
                        'solver': solver,
                        'penalty': penalty,
                        'C': C,
                        'accuracy': accuracy
                    })

    return results
```

```python
%%time
results = evaluate_logistic_regression(param_grid, X_train, X_test, y_train, y_test)
for result in results:
    print(result)
```

```python
def plot_accuracy_vs_C(results):
    c_values = [result['C'] for result in results]
    accuracies = [result['accuracy'] for result in results]
    solvers = [result['solver'] for result in results]

    plt.figure(figsize=(10, 6))

    for solver in set(solvers):
        x = [c_values[i] for i in range(len(c_values)) if solvers[i] == solver]
        y = [accuracies[i] for i in range(len(accuracies)) if solvers[i] == solver]
        if solver == 'saga':
            plt.plot(x, y, marker='o', linestyle=':', color='red', label='saga')
        elif solver == 'newton-cg':
            plt.plot(x, y, marker='s', linestyle='--', color='blue', label='newton-cg')
        elif solver == 'sag':
            plt.plot(x, y, marker='^', linestyle='-', color='green', label='sag')

    plt.xscale('log')
    plt.xlabel('C')
    plt.ylabel('Accuracy')
    plt.title('Accuracy vs C')
    plt.legend()
    plt.grid(True)
    plt.show()
```
